**Z-SCORE Normalization**

```python
data = [100,200,300,400,500,1000]
standard = np.std(data)
```

```python
standard
```

```python
mean = np.mean(data)
```

```python
means = (data - mean)/standard
```

```python
means
```

**MIN-MAX Normalization**

```python
min = np.min(data)
```

```python
max = np.max(data)
```

```python
minmax = (data - min)/(max - min)
```

```python
minmax
```

**Decimal Scaling**

```python
def decimal_scaling(data):
    max_value = np.max(np.abs(data))

    num_digits = len(str(int(max_value)))

    scaling_factor = 10 ** num_digits

    scaled_data = data / scaling_factor

    return scaled_data

data = np.array([100,200,300,400,500,1000])
scaled_data = decimal_scaling(data)
print("Original data:", data)
print("Scaled data:", scaled_data)
```

```python
data = np.array([200,300,400,600,1000])
scaled_data = decimal_scaling(data)
print("Original data:", data)
print("Scaled data:", scaled_data)
```

**1.Min-Max Normalization**

**1.1 Import the necessary library**

```python
from sklearn.preprocessing import MinMaxScaler
```

**Create the dataset**

```python
data = {
    'bedroom':[1,2,3,4,5],
    'bathroom':[1,2,2,3,2],
    'sqrt_living':[500,1000,1500,2000,2500],
    'age_of_the_house':[5,10,15,20,25],
    'price':[1000000,1500000,2000000,2500000,3000000]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

**Apply the Min_Max Normalization**

```python
scaler = MinMaxScaler()
```

**Fit**

```python
normalized_data = scaler.fit_transform(df)
```

```python
normalized_data
```

```python
normalized_data = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
```
